```python
def categorize_col(netflix_df, col, labels):

    edges = [netflix_df[col].describe()['min'],
            netflix_df[col].describe()['25%'],
            netflix_df[col].describe()['50%'],
            netflix_df[col].describe()['75%'],
            netflix_df[col].describe()['max']]
    netflix_df[col] = pd.cut(netflix_df[col], edges, labels=labels, duplicates='drop')
    return netflix_df
```

```python
labels = ['Not Popular', 'Below Average', 'Average', 'Popular']
categorize_col(netflix_df, 'Vote_Average', labels)

netflix_df['Vote_Average'].unique()
```

```python
netflix_df.dropna(inplace=True)
netflix_df.isna().sum()
```

```python
netflix_df['Genre'] = netflix_df['Genre'].str.split(', ')
netflix_df = netflix_df.explode('Genre').reset_index(drop=True)
netflix_df.head(10)
```

```python
netflix_df['Genre'] = netflix_df['Genre'].astype('category')
netflix_df['Genre'].dtypes
```

```python
netflix_df.info()
```

```python
sns.set_style('whitegrid')
```

```python
#Gnere Column Distribution

ax = sns.catplot(data=netflix_df, y='Genre', kind='count', order=netflix_df['Genre'].value_counts().index)

plt.title('Gnere Column Distribution');
plt.tight_layout()
plt.show()
```

**The majority of Netflix content falls under the Drama and Comedy genres, indicating these are the most popular categories on the platform. Genres such as Documentary and Western have the least representation, reflecting a much smaller audience or production focus for these types of content on Netflix.**

```python
#Vote Column Distribution

ax = sns.catplot(data=netflix_df, y='Vote_Average', kind='count', order=netflix_df['Vote_Average'].value_counts().index)

plt.title('Vote Distribution');
plt.tight_layout()
plt.show()
```

**The vote distribution for Netflix content is relatively balanced across all categories, with no single group significantly outweighing the others. This suggests a diverse range of viewer opinions, indicating that titles appeal to different audience segments with varying popularity and ratings.**
